```python
import fasttext
```

```python
model = fasttext.load_model('fastText/cc.ko.300.bin')
```

```python
model.get_dimension()
```

```python
model.get_word_vector('공부')
```

```python
model.get_word_vector('오늘은')
```

```python
from gensim import models

ko_model = models.fasttext.load_facebook_model('fastText/cc.ko.300.bin')
```

```python
ko_model.most_similar('유선전화', topn=10)
```

```python
ko_model.most_similar('기기변경', topn=10)
```
